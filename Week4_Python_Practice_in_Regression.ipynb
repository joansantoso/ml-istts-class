```python
y = df_new["price"].values
X = df_new.iloc[:,1:].values
```

```python
scaler_model.mean_
```

```python
y_scaled = y.reshape(-1,1)
```

```python
y_scaler_model = StandardScaler()
y = y_scaler_model.fit_transform(y_scaled).reshape(-1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size=0.3)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_hat = model.predict(X_test)
mse_score = mean_squared_error(y_test,y_hat)
print("MSE : ", mse_score)
```

```python
from sklearn.preprocessing import PolynomialFeatures

features_builder = PolynomialFeatures(degree=2, include_bias = True)
polynomial_features = features_builder.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(polynomial_features,y , test_size=0.3)
```
